#**Matriz fundamental**

La matriz fundamental es una herramienta de las cadenas de Markov que funciona como una especie de contador.

Toma el caos de las probabilidades de rebotar de un lado a otro en un tablero, y lo convierte en un mapa exacto de cuánto tiempo se va a pasar en cada lugar antes de que termine el recorrido.

Vamos a resolver los siguientes ejercicios

Ejercicio 1. ¿Cuál es el número promedio de tiradas necesarias para terminar el juego de serpientes y escaleras en el tablero de la imagen adjunta?


Captura de pantalla 2026-05-15 213059.png

**Resolucion analitica**

```python
# Importamos las librerias que vamos a usar
import numpy as np
import random
```

```python
P = np.zeros((21, 21)) # Aqui vemos la matriz de transicion de los estados del 0 al 20

camino = {3: 11, 13: 5, 15: 19, 17: 9} # Ponemos de donde sales y a donde llegan

for i in range(20):  # Ponemos el ciclo
    for dado in range(1, 7):
        siguiente = i + dado

        if siguiente > 20: # Hacemos que si se pasa de 20, se queda en el lugar
            siguiente = i
        else:
            if siguiente in camino: # Aqui si cae en serpiente o escalera se va a mover
                siguiente = camino[siguiente]

        P[i, siguiente] += 1/6

P[20, 20] = 1.0 # Vemos que el estado 20 es absorbente

Q = P[:20, :20] # Hacemos la matriz del estado 0 al 19
I = np.eye(20)

N = np.linalg.inv(I - Q) # Calculamos la matriz fundamental

t = np.dot(N, np.ones(20)) # El tiempo esperado

print("Ejercicio 1. Serpientes y escaleras")
print(f"Promedio de tiradas desde el inicio: {t[0]:.4f}")
```

**Resolucion por simulacion**

```python
def serpientes_escaleras(n=100000): # Definimos la funcion para la simulacion
    caminos = {3: 11, 13: 5, 15: 19, 17: 9}
    total_tiradas = 0

    for _ in range(n):
        posicion = 0
        tiradas = 0

        while posicion != 20:
            dado = random.randint(1, 6) # Aqui tiramos el dado
            tiradas += 1
            siguiente = posicion + dado

            if siguiente > 20:
                siguiente = posicion  # Si se pasa de 20, no se mueve
            else:
                if siguiente in caminos:  # Serpiente o escalera, se mueve
                    siguiente = caminos[siguiente]

            posicion = siguiente

        total_tiradas += tiradas

    return total_tiradas / n

simulacion = serpientes_escaleras(100000)

print("Resultados por simulacion")
print(f"Ejercicio 1. Serpientes y escaleras:")
print(f"Promedio simulado de tiradas: {simulacion:.4f}")
```

Ejercicio 2. ¿Cuál es la probabilidad de que el ratón, iniciando  en la casilla 0, alcance la comida?

**Resolucion analitica**

```python
adj = { # Conexiones entre los caminos
    0: [1, 2],
    1: [0, 3, 7],
    2: [0, 3, 8],
    3: [1, 2, 4, 5],
    4: [3, 6, 7],
    5: [3, 6, 8],
    6: [4, 5]
}

Q = np.zeros((7, 7)) # Matriz de transitorios por transitorios
R = np.zeros((7, 2)) # Matriz de transitorios por absorbentes

for i in range(7): # Usamos el ciclo
    vecinos = adj[i]
    grado = len(vecinos)
    prob = 1.0 / grado

    for vecino in vecinos:
        if vecino < 7:
            Q[i, vecino] = prob
        elif vecino == 7: # Comida
            R[i, 0] = prob
        elif vecino == 8: # Choque
            R[i, 1] = prob

I = np.eye(7)

N = np.linalg.inv(I - Q) # Calculamos la matriz fundamental

B = np.dot(N, R) # Calculamos las probabilidades

prob_comida_desde_0 = B[0, 0] # Esta parte es la probabilidad de empezar en 0 y terminar en 7

print("Ejercicio 2. Laberinto del raton")
print(f"Probabilidad de llegar a la comida desde la celda 0: {prob_comida_desde_0:.4f}")
```

**Resolucion por simulacion**

```python
def laberinto_raton(n=100000): # Definimos la funcion
    adj = { # Las conexiones del laberinto
        0: [1, 2],
        1: [0, 3, 7],
        2: [0, 3, 8],
        3: [1, 2, 4, 5],
        4: [3, 6, 7],
        5: [3, 6, 8],
        6: [4, 5]
    }

    exitos = 0 # Veces que el raton llega a la comida

    for _ in range(n):
        posicion = 0

        while posicion != 7 and posicion != 8: # Mientras no caiga en la comida ni en el choque
            posicion = random.choice(adj[posicion]) # Elige una puerta al azar

        if posicion == 7: # Si termina el ciclo porque llega a 7, es exito
            exitos += 1

    return exitos / n

sim_raton = laberinto_raton(100000)

print(f"Ejercicio 2. Laberinto del raton:")
print(f"Probabilidad simulada de llegar a la comida: {sim_raton:.4f}")
```
