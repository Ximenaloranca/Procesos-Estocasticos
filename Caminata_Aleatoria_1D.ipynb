# **Caminata Aleatoria 1D**

La caminata aleatoria representa el movimiento de una partícula o sistema que se desplaza en una línea recta, dando pasos discretos hacia la derecha o hacia la izquierda en función del azar

Tomamos una partícula que inicia en el origen $x_0 = 0$.

En cada paso de tiempo $t$, la partícula se mueve una unidad hacia la derecha o hacia la izquierda según las siguientes probabilidades:


*   Hacia la derecha ($+1$): con probabilidad $p$
*   Hacia la izquierda ($-1$): con probabilidad $q$, donde $q = 1 - p$.


```python
#Importamos las librerias que vamos a utilzar
import random as r
import matplotlib.pyplot as plt
```

Creamos la funcon para la caminata y usamos un ciclo for para delimitar las condiciones que se deben cumplir

```python
# Definimos la función de la caminata
def caminata(pasos, p):
    x = 0 # Lo usamos para la posicion inicial y agregamos la lista para guardar los datos
    h = []
    h.append(x) # lo agregamos para saber cual es el ultimo punto
# Se usa un ciclo for para generar un numero aleatorio y ademas agregamos las condiciones para que se cumpla, si es menor a p es exito y si es mayor a p es fracaso
    for i in range(pasos):
        u = r.random()
        if u < p:
            x += 1
            h.append(x)
        else:
            x -= 1
            h.append(x)

# Imprimimos ya el resultado
    print("La ultima posición: ", x)
    return h
```

Definimos los valores para comprobar que el codigo funcione y cumpla con las condiciones

```python
# Definimos los parametros
w = 100        # Número de pasos
p = 0.5         # Probabilidad de ir a la derecha

# Imprimimos ya la ultima posicion
trayectoria = caminata(w, p)
```

Sigue el codigo para la grafica

```python
# Hacemos la grafica
plt.figure(figsize=(12,6))
plt.axhline(y=0, linestyle="--", color='pink')
plt.title(f'Caminata aleatoria 1D (p={p})')
plt.xlabel('Iteraciones')
plt.ylabel('Valor')
plt.plot(trayectoria, color='orange')
plt.show()
```

**Conclusión:**

En este programa podemos ver como se acumulan muchisimos eventos independientes. Cuando la probabilidad p y la probabilidad q son exactamente de 0.5, el sistema no tiene una preferencia, pero la particula aun asi se mueve, de todas formas, podemos cambiarle los valores para seguir comprobando que esto funciona y cumple con las condiciones.

La caminata aleatoria de una dimension, tambien nos demuestra como el azar termina creando comportamientos y patrones que podemos estudiar y predecir a nivel macroscopico. El codigo que armamos funciona para dejarnos manipular las probabilidades de ir a un lado o al otro, y comprobar como esas reglas controlan el sistema a lo largo del tiempo.
