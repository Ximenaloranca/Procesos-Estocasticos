# **Tutorial para matrices**
En esta primera parte, se inclue un breve tutorial que explica como representar matrices, obtener una entrada dada (columna, renglón o entrada especifica), como multiplicar matrices, elevarlas a una potencia dada, como multiplicar vectores por matrices y hallar valores y vectores propios, usando la librería Sympy.


```python
# Importamos las librerias
import sympy as sp
import matplotlib.pyplot as plt

# Inicializar impresión bonita para SymPy
sp.init_printing()
```

En el siguiente codigo se representan matrices:

```python
A = sp.Matrix([[1, 2], # Creamos una matriz A de 2x2
               [3, 4]])
B = sp.Matrix([[2, 0],  # Creamos otra matriz B
               [1, 2]])

sp.pprint(A) # Las imprimimos
sp.pprint(B)
```

1.2 Obtener entradas, renglones y columnas
Recuerda que en Python los índices comienzan en cero.

```python
# 1. Entrada específica (renglón 0, columna 1)
entrada = A[0, 1]
print("Entrada A[0,1]:", entrada)

# 2. Obtener un renglón completo (el segundo renglón, índice 1)
renglon = A.row(1)
print("Segundo renglón:")
display(renglon)

# 3. Obtener una columna completa (la primera columna, índice 0)
columna = A.col(0)
print("Primera columna:")
display(columna)
```

1.3 Multiplicación de matrices y potencias
Las operaciones aritméticas estándar de Python funcionan de forma intuitiva.

```python
# Multiplicación de matrices (A * B)
C = A * B
print("A * B:")
display(C)

# Elevar una matriz a una potencia dada (A al cubo)
A_cubo = A**3
print("A^3:")
display(A_cubo)
```

1.4 Multiplicar vectores por matrices
Un vector columna en SymPy es simplemente una matriz con una sola columna.

```python
# Vector columna v
v = sp.Matrix([5, -1])

# Multiplicación A * v
Av = A * v
print("A * v:")
display(Av)
```

1.5 Valores y vectores propios
SymPy tiene funciones integradas para calcular la eigen-información de una matriz exacta.

```python
# Valores propios (Eigenvalues)
# Devuelve un diccionario: {valor_propio: multiplicidad_algebraica}
valores_propios = A.eigenvals()
print("Valores propios y sus multiplicidades:")
display(valores_propios)

# Vectores propios (Eigenvectors)
# Devuelve una lista de tuplas: (valor_propio, multiplicidad, [vectores_propios])
vectores_propios = A.eigenvects()
print("\nVectores propios:")
display(vectores_propios)
```

# **Tutorial diagrama de transicion**
En esta parte se explica como dibujar un diagrama de transición usando la librería networkx.

```python
# Importamos la libreria
import networkx as nx
```

```python
G=nx.DiGraph() #grafo dirigido
G.add_edge("A","B",capacity=2)
G.add_edge("B","C",capacity=2)
G.add_edge("B","D",capacity=8)
G.add_edge("A","C",capacity=3)
G.add_edge("C","D",capacity=8)
```

```python
G
nx.draw_networkx(G)
```

```python
nx.nodes(G) #Esta funcion devuelve la lista de nodos que conforman la red
```

```python
nx.edges(G)
```

```python
nx.number_of_nodes(G)
```
