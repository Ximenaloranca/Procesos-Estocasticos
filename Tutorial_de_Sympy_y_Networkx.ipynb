# **Tutorial para matrices**
En esta primera parte, se inclue un breve tutorial que explica como representar matrices, obtener una entrada dada (columna, rengl√≥n o entrada especifica), como multiplicar matrices, elevarlas a una potencia dada, como multiplicar vectores por matrices y hallar valores y vectores propios, usando la librer√≠a Sympy.


```python
# Importamos las librerias
import sympy as sp
import matplotlib.pyplot as plt
```

En el siguiente codigo se representan matrices:

```python
A = sp.Matrix([[1, 2], # Creamos una matriz A de 2x2
               [3, 4]])
B = sp.Matrix([[2, 0],  # Creamos otra matriz B
               [1, 2]])

sp.pprint(A) # Las imprimimos
sp.pprint(B)
```

Obtenemos las entradas, renglones y columnas

```python
entrada = A[0, 1] # Hacemos las entradas por rengl√≥n 0 y columna 1
print("Entrada A[0,1]:", entrada)

renglon = A.row(1) # Un rengl√≥n completo
print("Segundo rengl√≥n:")
sp.pprint(renglon)

columna = A.col(0) # Una columna completa
print("Primera columna:")
sp.pprint(columna)
```

En el siguiente codigo mltiplicamos matrices y hacemos potencias de las matrices

```python
C = A * B # Multiplicaci√≥n de la matriz A por la B
print("A por B:")
sp.pprint(C)

Acubo = A**3 # Elevamos la matriz al cubo o cualquier potencia
print("A al cubo:")
sp.pprint(Acubo)
```

En la siguiente parte se ultiplican vectores por matrices

```python
v = sp.Matrix([5, -1]) # Hacemos el vector columna

Av = A * v # Multiplicamos la matriz por el vector
print("A * v:")
sp.pprint(Av)
```

Aquie podemos ver los valores y vectores propios

```python
valores_propios = A.eigenvals() # Valores propios
print("Valores propios y multiplicidades:")
sp.pprint(valores_propios)

vectores_propios = A.eigenvects() # Los vectores propios y se devuelven el valor propio, la multiplicidad y los vectores propios
print("\nVectores propios:")
sp.pprint(vectores_propios)
```

# **Tutorial diagrama de transicion**
En esta parte se explica como dibujar un diagrama de transici√≥n usando la librer√≠a networkx.

```python
# Importamos la libreria
import networkx as nx
```

```python
G=nx.DiGraph() #grafo dirigido
G.add_edge("A","B",capacity=2)
G.add_edge("B","C",capacity=2)
G.add_edge("B","D",capacity=8)
G.add_edge("A","C",capacity=3)
G.add_edge("C","D",capacity=8)
```

```python
G
nx.draw_networkx(G)
```

```python
nx.nodes(G) #Esta funcion devuelve la lista de nodos que conforman la red
```

```python
nx.edges(G) # Me dice que direccion tiene el nodo, hacia que nodo va
```

```python
nx.number_of_nodes(G) # Me regresa el numero de nodos que hay
```
